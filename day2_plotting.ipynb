```python
import numpy as np
```

```python
run=np.array([100,50,91,78,89,25,34,19,9,10])
wic=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(run,wic)
plt.show()
```

```python
run=np.array([100,50,91,78,89,25,34,19,9,10])
wic=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(run,wic,color="purple",marker="^")
plt.show()
```

```python
st1=[23,45,67,89,90]
st2=[29,39,48,58,78]
sub=["math","sci","eng","tel","soc"]
plt.plot(sub,st1,label="st1",color="c")
plt.plot(sub,st2,label="st2",color="r")
plt.title("marks ")
plt.legend()
plt.show()
```

```python
st1=[23,45,67,89,90]
st2=[29,39,48,58,78]
sub=["math","sci","eng","tel","soc"]
plt.subplot(2,1,1)
plt.plot(sub,st1,label="student1",color="c")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,st2,label="student2",color="r")
plt.legend()
plt.show
```
